```python
# Task 1
def derivate(x, h):
    return (math.exp(x+h) - math.exp(x)) / h

true_value = math.exp(1.5)
# Finn ut hvor funksjonen starter å gi feil
print(true_value)
for i in range(1, 20):
    h = 10**(-i)
    print(h, derivate(1.5, h), abs(true_value - derivate(1.5, h)))
```

```python
# Task 2
def derivate2(x, h):
    return (math.exp(x+h) - math.exp(x-h)) / (2*h)

true_value = math.exp(1.5)
# Finn ut hvor funksjonen starter å gi feil
for i in range(1, 18):
    h = 10**(-i)
    print(h, derivate2(1.5, h), abs(true_value - derivate2(1.5, h)))
```

Man ser ved steg 1e-05 så var feilen lavest. Deretter stiger. Man ser at feilen minsker med 10², dette kan forklares ved Taylor rekke for f(x+h) og f(x-h). Når Taylor rekken settes inn for de to funksjonene så kanselleres mange ledd og vi ender opp med
$\frac{f(x+h)-f(x-h)}{2}=f'(x)+\frac{f'''(x)}{3!}h^2+O(h^4)$

Man ser på den approksimerte verdien av f'(x), som er det man ønsker å estimere, men det er også et tilleggsledd $\frac{f'''(x)}{3!}h^2$ som representerer feilen i tilnærmingen. Denne feilen ser man minker med $h^2$.

Grunnen til at feilen øker etterhvert er mest sannsynlig grunnet avrundingsfeil.

```python
# Task 3
def derivate3(x, h):
    return (-math.exp(x+2*h) + 8*math.exp(x+h) - 8*math.exp(x-h) + math.exp(x-2*h)) / (12*h)

true_value = math.exp(1.5)
# Finn ut hvor funksjonen starter å gi feil
for i in range(1, 18):
    h = 10**(-i)
    print(h, derivate3(1.5, h), abs(true_value - derivate3(1.5, h)))
```

```python
# Task 4 løs varmeligningen med eksplisitt skjema
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Kode for å velge numerisk metode
def heat_equation(h,k, boundaryCondition, intialConditions, time, space, scheme="explicit"):

    x = np.arange(0, space+h, h)
    t = np.arange(0, time+k, k)

    n = len(x)
    m = len(t)
    T = np.zeros((n, m))

    T[0, :] = boundaryCondition[0]
    T[-1, :] = boundaryCondition[1]

    T[:, 0] = intialConditions

    factor = k/(h**2)

    schemes = ["explicit", "implicit", "crank-nicolson"]
    if scheme not in schemes:
        raise ValueError("Invalid scheme")
    
    if scheme == "explicit":
        for j in range(1, m):
            for i in range(1, n-1):
                T[i, j] = factor*(T[i+1, j-1] - 2*T[i, j-1] + T[i-1, j-1]) + T[i, j-1]
    elif scheme == "implicit":
        A = np.diag([1+2*factor]*(n-2)) + np.diag([-factor]*(n-3), -1) + np.diag([-factor]*(n-3), 1)

        for j in range(1, m):
            b = T[1:-1, j-1].copy()
            b[0] = b[0] + factor*T[0, j]
            b[-1] = b[-1] + factor*T[-1, j]
            solution = np.linalg.solve(A, b)
            T[1:-1, j] = solution
    elif scheme == "crank-nicolson":
        A = np.diag([2+2*factor]*(n-2)) + np.diag([-factor]*(n-3), -1) + np.diag([-factor]*(n-3), 1)

        B = np.diag([2-2*factor]*(n-2)) + np.diag([factor]*(n-3), -1) + np.diag([factor]*(n-3), 1)
        for j in range(0, m-1):
            b = T[1:-1, j].copy()
            b = np.dot(B, b)
            b[0] = b[0] + factor*(T[0, j] + T[0, j+1])
            b[-1] = b[-1] + factor*(T[-1, j] + T[-1, j+1])
            solution = np.linalg.solve(A, b)
            T[1:-1, j+1] = solution

    plt.plot(x, T)
    plt.xlabel("Distance [m]")
    plt.ylabel("temperature [degree C]")
    plt.title(f"Scheme: {scheme}")
    plt.legend([f"t={i} s" for i in t], loc="upper right")
    plt.show()


h = 0.1
k = 0.025
time = 0.1
space = 1
x = np.arange(0, space+h, h)
t = np.arange(0, time+k, k)
boundaryCondition = [0, 0]
intialConditions = np.sin(np.pi*x)
heat_equation(h, k, boundaryCondition, intialConditions, time, space, scheme="explicit")
heat_equation(h, k, boundaryCondition, intialConditions, time, space, scheme="implicit")
heat_equation(h, k, boundaryCondition, intialConditions, time, space, scheme="crank-nicolson")
```
